### 평가가 이루어지지 않은 데이터 삭제

```python
# 총 데이터 - 1230376 
# 평점 0 - 684320
# 리뷰 0 - 684320
# 평점리뷰 모두 '0' - 684320
data = data[(data['Reviews'] != 0) & (data['Average_User_Rating'] != 0.0)]
data
```

```python
# 불필요 컬럼 제거
cols = ['App_Name', 'App_Id', 'AppStore_Url', 'Released', 'Updated', 'Required_IOS_Version', 'Version', 'DeveloperId', 'Developer', 'Developer_Url', 'Developer_Website', 'Current_Version_Score', 'Current_Version_Reviews', 'Currency']
data.drop(cols, axis=1, inplace=True)
```

```python
data
```

```python
data.rename(columns={'Primary_Genre' : 'Category',
                      'Content_Rating' : 'Available_Age',
                        'Size_Bytes' : 'Size',
                        'Average_User_Rating' : 'Rating'}, inplace=True)
data
```

```python
# null값확인
data.isnull().sum()
```

```python
# null값 채우기 - 평균값(trend반영X)
data['Price'].fillna(data['Price'].mean(), inplace=True)
data['Size'].fillna(data['Size'].mean(), inplace=True)
```

```python
data.isnull().sum()
```

```python
# 사이즈 MB화
data['Size'] = ((data['Size'] / (1024*1024)).round()).astype(int)
```

```python
# 가격 정수화, 평점 반올림
data['Price'] = data['Price'].round().astype(int)
data['Rating'] = data['Rating'].round(2)
```

```python
data['Category'].nunique(), data['Category'].unique()
```

```python
''' 
이용가능 연령대 - Integer Encoding
everyone - 0
kid - 1
children - 2
teens - 3
adults - 4
'''
data['Available_Age'] = data['Available_Age'].replace('4+', 1)
data['Available_Age'] = data['Available_Age'].replace('9+', 2)
data['Available_Age'] = data['Available_Age'].replace('12+', 3)
data['Available_Age'] = data['Available_Age'].replace('17+', 4)
data['Available_Age'] = data['Available_Age'].replace('Not yet rated', 0)
data['Available_Age'] = data['Available_Age'].astype(int)
```

```python
data
```

---

### Integer Encoding

```python
cate_list = data['Category'].unique()
cate_list
```

```python
cate_dict = {}
for index in range(len(cate_list)):
    cate_dict[cate_list[index]] = index

data['Cate_Int'] = data['Category'].map(cate_dict).astype(int)
data.head()
```

### One Hot Encoding

```python
data_dum = pd.get_dummies(data, columns=['Category'])
data_dum.columns
```

```python
data_dum
```
